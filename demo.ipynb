```python
import matplotlib.pyplot as plt
import keras_ocr
import cv2
import time
```

```python
# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images1 = [cv2.imread('./CBD_Images/test1.png'), cv2.imread('./CBD_Images/test2.png')]
images2 = [cv2.imread('./CBD_Images/test3.png'), cv2.imread('./CBD_Images/test4.png')]
images3 = [cv2.imread('./CBD_Images/test5.png'), cv2.imread('./CBD_Images/test6.png')]
images4 = [cv2.imread('./CBD_Images/test7.png'), cv2.imread('./CBD_Images/test8.png')]
images5 = [cv2.imread('./CBD_Images/test9.png'), cv2.imread('./CBD_Images/test10.png')]
images6 = [cv2.imread('./CBD_Images/test11.png'), cv2.imread('./CBD_Images/test12.png')]
```

```python
# Each list of predictions in prediction_groups is a list of
# (word, box) tuples.
images = images6
st = time.time()
prediction_groups = pipeline.recognize(images)
print("Inference Time: " + str(time.time()-st))
```

```python
# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(40, 40))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
plt.savefig("CBD6",
        dpi=None, 
        facecolor='w', 
        edgecolor='w',
        orientation='portrait', 
        papertype=None, 
        format=None,
        transparent=False, 
        bbox_inches=None, 
        pad_inches=0.1,
        frameon=None, 
        metadata=None)
```
